```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('sappelli_data\sappelli_results_clean.csv', sep=';')
```

```python
df.head()
df.shape
```

```python
# plot a histogram for each of the following features : Act1A1, Act2A1, ReasA1, REA1, NumTA1
import matplotlib.pyplot as plt

df[['Act1A1', 'Act2A1', 'ReasA1', 'REA1', 'NumTA1']].hist(bins=10)
plt.show()
```

```python
# Set the width of each bar
bar_width = 0.35

classes = ["request", "propose", "commit", "deliver", "amend", "refuse", "greet", "remind", "{}"]

# Set the x coordinates for the bars
x = np.arange(len(classes))

# Create subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Plot the first histogram
ax1.hist(df['Act1A1'], alpha=0.5, label='Act1A1')
ax1.set_xlabel('Class')
ax1.set_ylabel('Frequency')
ax1.set_title('Histogram of Act1A1')
ax1.set_xticks(x)
ax1.set_xticklabels(classes)
ax1.legend()

# Plot the second hist2ogram
ax2.hist(df['Act1A2'], bins=10, alpha=0.5, label='Act1A2')
ax2.set_xlabel('Class')
ax2.set_ylabel('Frequency')
ax2.set_title('Histogram of Act1A2')
ax2.set_xticks(x)
ax2.set_xticklabels(classes)
ax2.legend()

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# remove 6 rows with non numeric values in NumTA1 or NumTA2

df['NumTA1'] = pd.to_numeric(df['NumTA1'], errors='coerce')
df['NumTA2'] = pd.to_numeric(df['NumTA2'], errors='coerce')
df = df.dropna(subset=['NumTA1', 'NumTA2'])
df.shape
```

```python
# remove rows with {} in Act columns

df = df[~df['Act1A1'].str.contains('{}')]
df = df[~df['Act1A2'].str.contains('{}')]
df.shape
```
