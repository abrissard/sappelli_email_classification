```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

results = pd.read_csv('sappelli_data\sappelli_results_clean.csv', sep=';')
emails = pd.read_csv('sappelli_data\sappelli_emails.csv', sep=',')
```

```python
results.head()
```

```python
# rename column Id1 as mid
results = results.rename(columns={'Id1': 'mid'})
```

```python
emails.head()
```

```python
# merge data using mid and Id1 as keys
data = pd.merge(results, emails, on=['mid'], how='inner')
```

```python
data.head()
```

```python
# save data as csv
data.to_csv('sappelli_data\sappelli_data.csv', sep=';', index=False)
```
