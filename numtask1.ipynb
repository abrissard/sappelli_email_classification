```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('sappelli_data\sappelli_data.csv', sep=';')
```

```python
data.head()
```

```python
# Keep only NumTA1 column

numta1 = data[['NumTA1', 'body']]
numta1.head()
```

```python
numta1.describe()
```

```python
# remove rows with non numeric values in NumTA1

numta1 = numta1[numta1['NumTA1'].apply(lambda x: x.isnumeric())]
numta1.describe()
```

```python
# plot histogram of NumTA1

numta1['NumTA1'] = numta1['NumTA1'].astype(int)
numta1['NumTA1'].plot.hist(bins=100)
plt.show()
```

```python
# remove rows with 4 5 6

numta1 = numta1[numta1['NumTA1'] != 4]
numta1 = numta1[numta1['NumTA1'] != 5]
numta1 = numta1[numta1['NumTA1'] != 6]

numta1.describe()
```

```python
# plot histogram of NumTA1

numta1['NumTA1'] = numta1['NumTA1'].astype(int)
numta1['NumTA1'].plot.hist(bins=100)
plt.show()
```

```python
# save to csv

numta1.to_csv('sappelli_data\sappelli_data_numta1.csv', index=False)
```
