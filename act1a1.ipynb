```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('sappelli_data\sappelli_data.csv', sep=';')
```

```python
data.head()
```

```python
# Keep only NumTA1 column

act1a1 = data[['Act1A1', 'body']]
act1a1.head()
```

```python
act1a1.describe()
```

```python
# remove rows with values not in "request", "propose", "commit", "deliver", "amend", "refuse", "greet", "remind"

act1a1 = act1a1[act1a1['Act1A1'].isin(['request', 'propose', 'commit', 'deliver', 'amend', 'refuse', 'greet', 'remind'])]
act1a1.describe()
```

```python
# plot histogram of Act1A1

act1a1['Act1A1'].value_counts().plot(kind='bar')
plt.show()
```

```python
# save to csv

act1a1.to_csv('sappelli_data\sappelli_data_act1a1.csv', index=False)
```
